### Splitting features and target

```python
X=df.drop('target',axis=1)
X
```

```python
y=df.iloc[:,-1]
y
```

### Applying train test split for sending the data is training or testing data phase

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
clf=dt.fit(X_train,y_train)
```

```python
y_pred_train=clf.predict(X_train)
```

```python
y_pred_test=clf.predict(X_test)
```

### Classification Report

```python
from sklearn.metrics import classification_report,confusion_matrix,precision_recall_curve
```

### Confusion matrix

```python
print(confusion_matrix(y_pred_train,y_train))
```

```python
print(confusion_matrix(y_pred_test,y_test))
```

#### Here,the False Negative Rate is 10 

## As this dataset is of heart related serious issue we should focus more on False Negtive Rate.We will try to decrease False Negative rate to increase Recall of our model

```python
print(classification_report(y_pred_train,y_train))
```

```python
print(classification_report(y_pred_test,y_test))
```

## Hyper-Parameter Tuning

```python
param={'criterion':['gini','entropy'],
      'max_depth':np.arange(1,50),
      }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridclf=GridSearchCV(clf,param_grid=param,scoring='recall',cv=5,n_jobs=-1)
```

```python
gridclf
```

```python
gridclf.fit(X_train,y_train)
```

```python
gridclf.best_score_
```

```python
gridclf.best_params_
```

```python
grid_train_pred=gridclf.predict(X_train)
```

```python
grid_test_pred=gridclf.predict(X_test)
```

### Confusion Matrix of model after applying hyper-parameter tuning

```python
print(confusion_matrix(grid_train_pred,y_train))
```

```python
print(confusion_matrix(grid_test_pred,y_test))
```

### After applying hyper-parameter tuning False Negative Rate is 5

```python
print(classification_report(grid_test_pred,y_test))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier
```

### Splitting Features and target

```python
X=df.drop('target',axis=1)
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
```

### Fitting RandomForestClassifier model

```python
rf=RandomForestClassifier()
```

```python
clf=rf.fit(X_train,y_train)
```

```python
y_pred_train=clf.predict(X_train)
```

```python
y_pred_test=clf.predict(X_test)
```

### Confusion matrix

```python
print(confusion_matrix(y_pred_train,y_train))
```

```python
print(confusion_matrix(y_pred_test,y_test))
```

### Classification Report

```python
print(classification_report(y_pred_train,y_train))
```

```python
print(classification_report(y_pred_test,y_test))
```

### Hyper-Parameter Tuning

```python
param_grid={'criterion':['gini','entropy'],
            'class_weight':[None,'balanced'],
            'n_estimators':[100,150,300,600,500],
            'max_depth':np.arange(1,50,2)
           }
```

```python
gridclf=GridSearchCV(clf,param_grid=param_grid,scoring='recall',cv=5,n_jobs=-1)
```

```python
gridclf.fit(X_train,y_train)
```

```python
gridclf.best_score_
```

```python
gridclf.best_params_
```

```python
grid_train_pred=gridclf.predict(X_train)
```

```python
grid_test_pred=gridclf.predict(X_test)
```

### Confusion Matrix

```python
print(confusion_matrix(grid_train_pred,y_train))
```

```python
print(confusion_matrix(grid_test_pred,y_test))
```

### Here False Negative Rate is 4

## Clasification Report

```python
print(classification_report(grid_train_pred,y_train))
```

```python
print(classification_report(grid_test_pred,y_test))
```
