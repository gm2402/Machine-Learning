```python
X=df.drop('diagnosis',axis=1)
```

```python
X
```

```python
y=df['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
print('Train Data',X_train.shape,y_train.shape)
```

```python
print('Test Data',X_test.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import classification_report
```

```python
def my_model(model):
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)
    print('Train Data')
    print(classification_report(y_train,y_pred_train))
    print('Test Data')
    print(classification_report(y_test,y_pred_test))
```

```python
rf=RandomForestClassifier()
```

```python
my_model(rf)
```

```python
#n_estimators 
```

```python
rf1=RandomForestClassifier(n_estimators=200,max_depth=10,criterion='entropy')
```

```python
my_model(rf1)
```

```python
rf2=RandomForestClassifier(n_estimators=500,max_depth=12,criterion='entropy')
```

```python
my_model(rf2)
```

```python
param_grid={'criterion':['gini','entropy'],
            'class_weight':[None,'balanced'],
            'n_estimators':[100,150,300,600,500],
            'max_depth':np.arange(1,50,2),
            'min_samples_leaf':np.arange(1,20),
            'min_samples_split':np.arange(2,15,1)
           }
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
clf=RandomizedSearchCV(rf,param_distributions=param_grid,n_iter=10,scoring='f1',n_jobs=-1)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
my_model(clf)
```
