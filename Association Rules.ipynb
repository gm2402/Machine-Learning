```python
import pandas as pd
```

```python
df=pd.read_csv("GroceryStoreDataSet.csv",header=None,names=['Product'])
```

```python
df.head()
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
def break_data(x):
  return x.split(",")
```

```python
data=list(df['Product'].map(break_data))
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
encoded_data=te.fit_transform(data)
```

```python
encoded_data
```

```python
te.columns_
```

```python
df=pd.DataFrame(encoded_data,columns=te.columns_)
```

```python
df.head()
```

```python
df.replace(False,0,inplace=True)
```

```python
df.replace(True,1,inplace=True)
```

```python
df.head()
```

```python
scores=apriori(df,min_support=0.20,use_colnames=True)
scores
```

```python
rules=association_rules(scores,min_threshold=0.5)
rules
```
